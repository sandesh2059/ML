### why choose matplotlib and seaborn

Both librarues are anong the nost popular choices for data visualization in python. Here's why:
Matplotib: Known for its flexiblity, Matplotlib allows detailed customization of visual elements, making it perfect for tailored plots.

seaborn: Built on top of Matplotlib, Seaborn simplifies visualization tasks by offering high-level functions to create aesthetically pleasing statistical graphics.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# basics of matplotlib
#line plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, label="Sine wave", color="green")
plt.title("line plot example")
plt.xlabel("X=axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
# Bar chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 9, 20, 12, 25]
plt.title("Bar chart example")
color = ["green", "red", "blue", "pink", "yellow"]
plt.bar(categories, values, color=color)
plt.xlabel("categories")
plt.ylabel("values")
plt.show()
```

### visualization with seaborn

```python
import seaborn as sns
```

```python
#histogram and KDE plot:
data = np.random.randn(1000)
sns.histplot(data, kde=True, color='green')
plt.title("HISTOGRAN with KDE")
plt.show()
```

```python
#heatmap
matrix = np.random.rand(5, 5)
matrix
sns.heatmap(matrix, annot=True, cmap="magma")
plt.title("HEATMAP EXAMPLE")
plt.show()
```

```python
#pairplot example
sns.set_style("whitegrid")
df = sns.load_dataset("iris")
sns.pairplot(df, hue="species", markers = ["o", "s", "D"])
plt.show()
```
